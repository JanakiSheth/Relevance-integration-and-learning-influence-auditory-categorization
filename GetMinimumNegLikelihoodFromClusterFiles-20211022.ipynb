```python
import pdb
import numpy as np
import matplotlib.pyplot as plt

def strip_new_line_char(string):
    ""
    if "\n" in string:
        return string[:-1]
    return string
```

```python
subject = 'c746'

sensoryStd = 0.29
subjectArray = np.zeros((100,7))

file_parameters = open("auditory_categorization_prolific_online_data/preprocess_results/90-3000_100%_logPerceptAndLogLatent_GoFrom-InfToInf/results_from_cluster/resultsProlificData/errorbarsOnRelevanceMetric/"+subject+'.txt',"w")
for iterations in range(1,101):
    file_dev = "auditory_categorization_prolific_online_data/preprocess_results/90-3000_100%_logPerceptAndLogLatent_GoFrom-InfToInf/results_from_cluster/resultsProlificData/ControlResults/subsampledData/results_"+subject+'_'+str(iterations)+'.txt'
    with open(file_dev, 'r') as f:
        filelines = f.readlines()   
    fileArray = np.zeros((1,7))
    iline = 0
    while iline < len(filelines):
        if filelines[iline].startswith('[') and filelines[iline].endswith(']\n'):
            lineWithNewLineChar = strip_new_line_char(filelines[iline])
            lineElements = lineWithNewLineChar.split(",")
            temp = np.zeros((1,7))
            temp[:,6] = float(lineElements[6][:-1])
            temp[:,5] = float(lineElements[5][:-2])
            temp[:,4] = float(lineElements[4][:])
            temp[:,3] = float(lineElements[3][:])
            temp[:,2] = float(lineElements[2][:])
            temp[:,1] = float(lineElements[1][:])
            temp[:,0] = float(lineElements[0][8:])
            fileArray = np.append(fileArray,temp,axis=0)
            #print(fileArray[:,iline], iline)
            #pdb.set_trace()
            iline += 1
        else:
            line1WithNewLineChar = strip_new_line_char(filelines[iline])
            lineElementsOfFirstLine = line1WithNewLineChar.split(",")
            line2WithNewLineChar = strip_new_line_char(filelines[iline+1])
            lineElementsOfSecondLine = line2WithNewLineChar.split(",")
            lineElements = lineElementsOfFirstLine[:-1] + lineElementsOfSecondLine
            temp = np.zeros((1,7))
            temp[:,6] = float(lineElements[6][:-1])
            temp[:,5] = float(lineElements[5][:-2])
            temp[:,4] = float(lineElements[4][:])
            temp[:,3] = float(lineElements[3][:])
            temp[:,2] = float(lineElements[2][:])
            temp[:,1] = float(lineElements[1][:])
            temp[:,0] = float(lineElements[0][8:])
            #print(line1WithNewLineChar, line2WithNewLineChar, lineElements)
            #print(fileArray[:,iline], iline)
            fileArray = np.append(fileArray,temp,axis=0)
            iline += 2 
    
    fileArraySSConstrained_upperlimit = fileArray[fileArray[:,3]<sensoryStd+0.03]
    fileArraySSConstrained_lowerlimit = fileArraySSConstrained_upperlimit[fileArraySSConstrained_upperlimit[:,3]>sensoryStd-0.03]
    subjectArray[iterations-1,:] = fileArraySSConstrained_lowerlimit[np.argmin(fileArraySSConstrained_lowerlimit[:,-1])]
    file_parameters.write(str(np.around(subjectArray[iterations-1,:-1],decimals=2))+'\n')
    print(np.around(subjectArray[iterations-1,:-1],decimals=2))
    #pdb.set_trace()
```

```python
print("Mean",np.around(np.mean(subjectArray[:iterations-1],axis=0),decimals=2))
print("Std",np.around(np.std(subjectArray[:iterations-1],axis=0),decimals=2))
print("Median pback",np.around(np.quantile(subjectArray[:iterations-1,4],q=0.5),decimals=2))
print("5th quantile pback",np.around(np.quantile(subjectArray[:iterations-1,4],q=0.05),decimals=2))
print("95th quantile pback",np.around(np.quantile(subjectArray[:iterations-1,4],q=0.95),decimals=2))
print("Median LL",np.around(np.quantile(subjectArray[:iterations-1,-1]*6/6,q=0.5),decimals=2))
print("5th quantile LL",np.around(np.quantile(subjectArray[:iterations-1,-1]*6/6,q=0.05),decimals=2))
print("95th quantile LL",np.around(np.quantile(subjectArray[:iterations-1,-1]*6/6,q=0.95),decimals=2))
```
